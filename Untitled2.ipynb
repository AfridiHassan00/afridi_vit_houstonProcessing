```python
# Cell 1: Import Libraries and Load Data
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
import scipy.io as sio
import os
import time

def analyze_mat_file(data, name):
    """
    Analyze the contents and structure of a .mat file
    """
    print(f"\n{name} Analysis:")
    keys = [k for k in data.keys() if not k.startswith('__')]
    print(f"Keys: {keys}")
    
    for key in keys:
        array = data[key]
        print(f"\nKey '{key}':")
        print(f"Shape: {array.shape}")
        print(f"Data type: {array.dtype}")
        if array.size > 0:
            print(f"Value range: [{np.min(array)}, {np.max(array)}]")

def load_mat_files(data_dir):
    """
    Load and analyze the Houston dataset .mat files
    """
    print("Loading and analyzing .mat files...")
    
    hsi_data = sio.loadmat(os.path.join(data_dir, 'houston_hsi.mat'))
    gt_data = sio.loadmat(os.path.join(data_dir, 'houston_gt.mat'))
    index_data = sio.loadmat(os.path.join(data_dir, 'houston_index.mat'))
    
    analyze_mat_file(hsi_data, "Hyperspectral Image (HSI)")
    analyze_mat_file(gt_data, "Ground Truth (GT)")
    analyze_mat_file(index_data, "Index Data")
    
    return {
        'hsi_data': hsi_data,
        'gt_data': gt_data,
        'index_data': index_data
    }

# Load and analyze the data
data_dir = '.'  # Change this to your local data directory
dataset = load_mat_files(data_dir)

# Extract and display HSI dimensions
hsi = dataset['hsi_data']['houston_hsi'] if 'houston_hsi' in dataset['hsi_data'] else None
if hsi is not None:
    print("\nHSI Data Summary:")
    print(f"Shape: {hsi.shape}")
    print(f"Number of channels: {hsi.shape[2]}")
    print(f"Spatial resolution: {hsi.shape[0]}×{hsi.shape[1]}")
```

```python
# Cell 2: Extract Data and Analyze Ground Truth
# Extract data from loaded files
hsi = dataset['hsi_data']['houston_hsi']
gt = dataset['gt_data']['houston_gt']
index_data = dataset['index_data']

# Analyze ground truth classes
unique_classes = np.unique(gt)
unique_classes = unique_classes[unique_classes != 0]  # Remove background class if present

print("Ground Truth Analysis:")
print(f"Number of classes: {len(unique_classes)}")
print("Class labels:", unique_classes)

# Count samples per class
class_counts = {}
for cls in unique_classes:
    count = np.sum(gt == cls)
    class_counts[cls] = count
    print(f"Class {cls}: {count} pixels")

# Visualize ground truth
plt.figure(figsize=(10, 5))
plt.imshow(gt)
plt.colorbar(label='Class')
plt.title('Ground Truth Map')
plt.show()

print("\nHSI Data Statistics:")
print(f"Min value: {hsi.min():.2f}")
print(f"Max value: {hsi.max():.2f}")
print(f"Mean value: {hsi.mean():.2f}")
print(f"Std value: {hsi.std():.2f}")
```

```python
# Cell 3: Analyze Train-Test Split
def analyze_indices(index_data):
    """
    Analyze the train-test split indices
    """
    # Find the keys containing train/test indices
    train_keys = [k for k in index_data.keys() if 'train' in k.lower()]
    test_keys = [k for k in index_data.keys() if 'test' in k.lower()]
    
    print("Index file analysis:")
    print(f"Train index keys found: {train_keys}")
    print(f"Test index keys found: {test_keys}")
    
    # Extract indices
    train_indices = None
    test_indices = None
    
    if train_keys and test_keys:
        train_indices = index_data[train_keys[0]]
        test_indices = index_data[test_keys[0]]
        
    return train_indices, test_indices

# Get train-test indices
train_indices, test_indices = analyze_indices(index_data)

if train_indices is not None and test_indices is not None:
    print("\nTrain-Test Split Analysis:")
    print(f"Number of training samples: {len(train_indices)}")
    print(f"Number of testing samples: {len(test_indices)}")
    
    # Visualize train-test split
    split_map = np.zeros_like(gt)
    train_coords = np.unravel_index(train_indices.flatten(), gt.shape)
    test_coords = np.unravel_index(test_indices.flatten(), gt.shape)
    
    split_map[train_coords] = 1
    split_map[test_coords] = 2
    
    plt.figure(figsize=(10, 5))
    plt.imshow(split_map)
    plt.colorbar(ticks=[0, 1, 2], label='0: Unlabeled, 1: Train, 2: Test')
    plt.title('Train-Test Split Visualization')
    plt.show()
```

```python
# # Cell 5: Split Patches into Train and Test Sets
# def split_patches(patches, labels, coordinates, train_indices, test_indices):
#     """
#     Split patches into training and testing sets based on provided indices
#     """
#     # Create a map from coordinates to patch indices
#     coord_to_idx = {(i, j): idx for idx, (i, j) in enumerate(coordinates)}
    
#     # Get train and test coordinates
#     train_coords = np.unravel_index(train_indices.flatten(), gt.shape)
#     test_coords = np.unravel_index(test_indices.flatten(), gt.shape)
    
#     # Get indices for train and test patches
#     train_patch_indices = []
#     test_patch_indices = []
    
#     for i, j in zip(*train_coords):
#         if (i, j) in coord_to_idx:
#             train_patch_indices.append(coord_to_idx[(i, j)])
            
#     for i, j in zip(*test_coords):
#         if (i, j) in coord_to_idx:
#             test_patch_indices.append(coord_to_idx[(i, j)])
    
#     # Split the data
#     X_train = patches[train_patch_indices]
#     y_train = labels[train_patch_indices]
#     X_test = patches[test_patch_indices]
#     y_test = labels[test_patch_indices]
    
#     return X_train, y_train, X_test, y_test

# # Split the patches
# X_train, y_train, X_test, y_test = split_patches(patches, labels, coordinates, 
#                                                 train_indices, test_indices)

# print("Dataset splits:")
# print(f"Training set shape: {X_train.shape}")
# print(f"Training labels shape: {y_train.shape}")
# print(f"Testing set shape: {X_test.shape}")
# print(f"Testing labels shape: {y_test.shape}")

# # Verify class distribution
# print("\nClass distribution in training set:")
# train_classes, train_counts = np.unique(y_train, return_counts=True)
# for cls, count in zip(train_classes, train_counts):
#     print(f"Class {cls}: {count} samples")

# print("\nClass distribution in testing set:")
# test_classes, test_counts = np.unique(y_test, return_counts=True)
# for cls, count in zip(test_classes, test_counts):
#     print(f"Class {cls}: {count} samples")

# # Cell 5 (Modified): Split Patches into Train and Test Sets
# def split_patches(patches, labels, coordinates, train_indices, test_indices):
#     """
#     Split patches into training and testing sets based on provided indices
#     """
#     # Create coordinate to index mapping
#     coord_to_idx = {(i, j): idx for idx, (i, j) in enumerate(coordinates)}
    
#     # Get train and test coordinates
#     train_coords = np.unravel_index(train_indices.flatten(), gt.shape)
#     test_coords = np.unravel_index(test_indices.flatten(), gt.shape)
    
#     # Get indices for train and test patches
#     train_patch_indices = [coord_to_idx[(i, j)] for i, j in zip(*train_coords) if (i, j) in coord_to_idx]
#     test_patch_indices = [coord_to_idx[(i, j)] for i, j in zip(*test_coords) if (i, j) in coord_to_idx]
    
#     # Split the data
#     X_train = patches[train_patch_indices]
#     y_train = labels[train_patch_indices]
#     X_test = patches[test_patch_indices]
#     y_test = labels[test_patch_indices]
    
#     # Verify class distribution
#     for cls in range(1, 21):  # 20 classes
#         train_count = np.sum(y_train == cls)
#         test_count = np.sum(y_test == cls)
#         print(f"Class {cls}: Train={train_count}, Test={test_count}")
    
#     return X_train, y_train, X_test, y_test

# Cell 5 (Modified): Split Patches into Train and Test Sets
def split_patches(patches, labels, coordinates, train_indices, test_indices):
    """
    Split patches into training and testing sets based on provided indices
    """
    # Create coordinate to index mapping
    coord_to_idx = {(i, j): idx for idx, (i, j) in enumerate(coordinates)}
    
    # Get train and test coordinates
    train_coords = np.unravel_index(train_indices.flatten(), gt.shape)
    test_coords = np.unravel_index(test_indices.flatten(), gt.shape)
    
    # Get indices for train and test patches
    train_patch_indices = [coord_to_idx[(i, j)] for i, j in zip(*train_coords) if (i, j) in coord_to_idx]
    test_patch_indices = [coord_to_idx[(i, j)] for i, j in zip(*test_coords) if (i, j) in coord_to_idx]
    
    # Split the data
    X_train = patches[train_patch_indices]
    y_train = labels[train_patch_indices]
    X_test = patches[test_patch_indices]
    y_test = labels[test_patch_indices]
    
    print("Dataset splits:")
    print(f"Training set shape: {X_train.shape}")
    print(f"Training labels shape: {y_train.shape}")
    print(f"Testing set shape: {X_test.shape}")
    print(f"Testing labels shape: {y_test.shape}")
    
    print("\nClass distribution in training set:")
    train_classes, train_counts = np.unique(y_train, return_counts=True)
    for cls, count in zip(train_classes, train_counts):
        print(f"Class {cls}: {count} samples")
    
    print("\nClass distribution in testing set:")
    test_classes, test_counts = np.unique(y_test, return_counts=True)
    for cls, count in zip(test_classes, test_counts):
        print(f"Class {cls}: {count} samples")
    
    # Print comparison with expected values from the paper
    print("\nComparison with expected values from paper:")
    expected_train = {1: 1000, 2: 1000, 3: 1000, 4: 1000, 5: 1000, 
                     6: 1000, 7: 500, 8: 1000, 9: 1000, 10: 1000,
                     11: 1000, 12: 1000, 13: 1000, 14: 1000, 15: 1000,
                     16: 1000, 17: 250, 18: 1000, 19: 1000, 20: 1000}
    
    for cls in range(1, 21):
        train_count = np.sum(y_train == cls)
        test_count = np.sum(y_test == cls)
        expected_train_count = expected_train.get(cls, 0)
        print(f"Class {cls}:")
        print(f"  Train - Expected: {expected_train_count}, Got: {train_count}")
        print(f"  Test  - Got: {test_count}")
    
    return X_train, y_train, X_test, y_test

# Split the patches
X_train, y_train, X_test, y_test = split_patches(patches, labels, coordinates, 
                                                train_indices, test_indices)
```

```python
# Cell 6: Convert to PyTorch Tensors and Normalize
# Convert to PyTorch tensors
X_train_tensor = torch.from_numpy(X_train).float()
y_train_tensor = torch.from_numpy(y_train).long()
X_test_tensor = torch.from_numpy(X_test).float()
y_test_tensor = torch.from_numpy(y_test).long()

# Normalize the data
# Calculate mean and std on training set
mean = X_train_tensor.mean(dim=(0, 2, 3), keepdim=True)
std = X_train_tensor.std(dim=(0, 2, 3), keepdim=True)

# Avoid division by zero
std[std == 0] = 1.0

# Normalize training and testing data
X_train_tensor_normalized = (X_train_tensor - mean) / std
X_test_tensor_normalized = (X_test_tensor - mean) / std

print("Tensor shapes after normalization:")
print(f"X_train_tensor: {X_train_tensor_normalized.shape}")
print(f"y_train_tensor: {y_train_tensor.shape}")
print(f"X_test_tensor: {X_test_tensor_normalized.shape}")
print(f"y_test_tensor: {y_test_tensor.shape}")

print("\nNormalization statistics:")
print(f"Mean shape: {mean.shape}")
print(f"Std shape: {std.shape}")
print(f"Min value after normalization: {X_train_tensor_normalized.min().item():.4f}")
print(f"Max value after normalization: {X_train_tensor_normalized.max().item():.4f}")
```

```python
# Cell 7: Create PyTorch Dataset Class
class HyperspectralPatchDataset(Dataset):
    """PyTorch Dataset for hyperspectral patches"""
    
    def __init__(self, patches, labels):
        self.patches = patches
        self.labels = labels
        
    def __len__(self):
        return len(self.patches)
    
    def __getitem__(self, idx):
        return self.patches[idx], self.labels[idx]

# Create datasets
train_dataset = HyperspectralPatchDataset(X_train_tensor_normalized, y_train_tensor)
test_dataset = HyperspectralPatchDataset(X_test_tensor_normalized, y_test_tensor)

# Create data loaders
batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"Number of training batches: {len(train_loader)}")
print(f"Number of testing batches: {len(test_loader)}")

# Verify batch shapes
for patches, labels in train_loader:
    print(f"\nSample batch shapes:")
    print(f"Patches shape: {patches.shape}")
    print(f"Labels shape: {labels.shape}")
    break
```

```python
# Cell 8: Visualize Sample Patches
def plot_sample_patches(patches, labels, num_samples=5):
    """Plot sample patches and their labels"""
    if patches.shape[0] < num_samples:
        num_samples = patches.shape[0]
    
    fig, axes = plt.subplots(num_samples, 1, figsize=(8, 3*num_samples))
    if num_samples == 1:
        axes = [axes]
    
    for i in range(num_samples):
        # Get a random sample
        idx = np.random.randint(0, patches.shape[0])
        patch = patches[idx]
        label = labels[idx]
        
        # For visualization, use first 3 channels or single channel
        if patch.shape[0] >= 3:
            # Use first 3 channels as RGB
            rgb = patch[:3].permute(1, 2, 0).cpu().numpy()
            # Normalize to [0, 1] for display
            rgb_min = rgb.min()
            rgb_max = rgb.max()
            if rgb_max > rgb_min:
                rgb = (rgb - rgb_min) / (rgb_max - rgb_min)
        else:
            # Use first channel as grayscale
            rgb = patch[0].cpu().numpy()
        
        axes[i].imshow(rgb)
        axes[i].set_title(f"Class: {label.item()}")
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

print("Sample Training Patches:")
plot_sample_patches(X_train_tensor_normalized, y_train_tensor)

print("\nSample Testing Patches:")
plot_sample_patches(X_test_tensor_normalized, y_test_tensor)
```

```python
# Cell 9: Save the Processed Tensors
def save_tensors(save_dir):
    """Save the processed tensors to local directory"""
    # Create directory if it doesn't exist
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    # Save tensors
    torch.save(X_train_tensor_normalized, os.path.join(save_dir, 'X_train.pt'))
    torch.save(y_train_tensor, os.path.join(save_dir, 'y_train.pt'))
    torch.save(X_test_tensor_normalized, os.path.join(save_dir, 'X_test.pt'))
    torch.save(y_test_tensor, os.path.join(save_dir, 'y_test.pt'))
    
    # Save normalization parameters
    norm_params = {
        'mean': mean,
        'std': std
    }
    torch.save(norm_params, os.path.join(save_dir, 'normalization_params.pt'))
    
    print(f"\nTensors saved to '{save_dir}'")
    print("\nSummary of saved tensors:")
    print(f"X_train: {X_train_tensor_normalized.shape}, dtype={X_train_tensor_normalized.dtype}")
    print(f"y_train: {y_train_tensor.shape}, dtype={y_train_tensor.dtype}")
    print(f"X_test: {X_test_tensor_normalized.shape}, dtype={X_test_tensor_normalized.dtype}")
    print(f"y_test: {y_test_tensor.shape}, dtype={y_test_tensor.dtype}")

# Save the tensors
save_dir = './tensors'  # Change this to your desired local save directory
save_tensors(save_dir)
```

```python
# Cell 10: Function to Load Saved Tensors
def load_tensors(base_dir='./tensors'):
    """
    Load the saved tensors from disk
    
    Args:
        base_dir: Directory containing the saved tensors
        
    Returns:
        X_train, y_train, X_test, y_test: PyTorch tensors
    """
    X_train = torch.load(os.path.join(base_dir, 'X_train.pt'))
    y_train = torch.load(os.path.join(base_dir, 'y_train.pt'))
    X_test = torch.load(os.path.join(base_dir, 'X_test.pt'))
    y_test = torch.load(os.path.join(base_dir, 'y_test.pt'))
    
    try:
        norm_params = torch.load(os.path.join(base_dir, 'normalization_params.pt'))
        mean = norm_params['mean']
        std = norm_params['std']
        print("Normalization parameters loaded")
    except:
        print("No normalization parameters found")
    
    print(f"\nLoaded tensors with shapes:")
    print(f"X_train: {X_train.shape}")
    print(f"y_train: {y_train.shape}")
    print(f"X_test: {X_test.shape}")
    print(f"y_test: {y_test.shape}")
    
    return X_train, y_train, X_test, y_test

# Verify loading
print("Loading the saved tensors for verification:")
X_train_loaded, y_train_loaded, X_test_loaded, y_test_loaded = load_tensors()

# Verify they match
print("\nVerifying tensor equality:")
print(f"X_train matches: {torch.all(X_train_tensor_normalized == X_train_loaded)}")
print(f"y_train matches: {torch.all(y_train_tensor == y_train_loaded)}")
print(f"X_test matches: {torch.all(X_test_tensor_normalized == X_test_loaded)}")
print(f"y_test matches: {torch.all(y_test_tensor == y_test_loaded)}")
```
